Data frames in pandas

```python
import pandas as pd
import numpy as np
from numpy.random import randn
myData = randn(4,3) # 4 rows and 3 columns.
myRows = ("A", "B", "C", "D")
myColumns = ("Col1", "Col2", "Col3")

#creating the dataframe.
myDataframe = pd.DataFrame(myData, myRows, myColumns)
print(myDataframe)
```
